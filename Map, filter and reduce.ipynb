```python
def double(n):
    return n*2
```

```python
#double each number in the list
numbers = [1,2,3,4,5,6,7]
output = []
for num in numbers:
    output.append(double(num))

print(output)
```

```python
map(function, iterable) 

applies a given function to every item of an iterable
it returns new iterable with transformed values
```

```python
result = list(map(double, numbers))
```

```python
result
```

```python
list(map(lambda x: x*2, numbers))
```

```python
temp_clesuis = (10, 20, 30, 40, 50)
def clecius_to_fahrenheit(c):
    print((c * 9/5) + 32)

tuple(map(clecius_to_fahrenheit, temp_clesuis))
```

```python
#find even numbers
```

```python
def even_odd(n):
    if n%2 == 0:
        return True
    else:
        return False
```

```python
numbers = [1,2,3,4,5,6,7]
output = []
for n in numbers:
    if even_odd(n):
        output.append(n)
```

```python
output
```

```python
#filter(function, iterable)
list(filter(even_odd, numbers))
```

```python
from functools import reduce
```

```python
reduce(function, iterable)

#cumulative opeartion
takes 2 elements at a time, applies the function and carry forward it to next pair
```

```python
numbers = [1,2,3,4,5,6,7]
```

```python
#without using reduce
total = 0
for num in numbers:
    total += num

print(total)
```

```python
reduce(lambda x, y: x + y, numbers)
```
